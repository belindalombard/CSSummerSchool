# üîÅ Monte Carlo Madness: Estimating œÄ with Randomness

Welcome to a simple but powerful demo of **Monte Carlo methods** ‚Äî a technique that uses randomness to solve problems that might be deterministic.

In this notebook, you'll estimate the value of œÄ using **random sampling**.

You'll:
- Run a simulation that drops random points into a square
- Count how many fall inside a quarter circle
- Use that to estimate œÄ!


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Function to run the simulation
def estimate_pi(num_points=1000):
    x = np.random.uniform(0, 1, num_points)
    y = np.random.uniform(0, 1, num_points)
    inside = x**2 + y**2 <= 1
    pi_estimate = 4 * np.sum(inside) / num_points

    theta = np.linspace(0, np.pi/2, 300)
    circle_x = np.cos(theta)
    circle_y = np.sin(theta)

    plt.figure(figsize=(6, 6))
    plt.scatter(x[inside], y[inside], color='blue', s=1, label='Inside Circle')
    plt.scatter(x[~inside], y[~inside], color='red', s=1, label='Outside Circle')
    plt.plot(circle_x, circle_y, 'k--', linewidth=1.5, label='Quarter Circle')
    plt.title(f'Estimate of œÄ = {pi_estimate:.4f} with {num_points} points')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.axis('square')
    plt.grid(True)
    plt.show()

    return pi_estimate
```

```python
# Try changing this number!
num_points = 100
#

pi_estimate = estimate_pi(num_points=num_points)

print(f'Pi estimate: {pi_estimate}')
print(f'Actual value of pi: {np.pi}')
```
